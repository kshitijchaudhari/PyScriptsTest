```python
import matplotlib.pyplot as plt
```

```python
y = [1, 2, 5, 12,14,6,4]
x = [3, 5, 7, 5, 3,2 ,6]
```

```python
fig = plt.figure(figsize=(15, 7))
plt.plot(x, y)
plt.xlim(0,15)
plt.ylim(0,15)
plt.xlabel('Count')
plt.ylabel('Wins')
plt.title('Wins over count graph')
plt.suptitle('Sample Matplotlib Pyplot')
```
